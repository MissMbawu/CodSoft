# Data Cleaning

```python
# Missing Values

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
correlation_mat = df.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# High Correlation

high_corr = corr_pairs[(corr_pairs) > 0.5]
high_corr
```
